```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_gender_associtation(llm_model):
    load_path_personX =  f'../results/gender_association/{llm_model}/{llm_model}_PersonX_gender_association_ALL.csv'
    load_path_unisex =  f'../results/gender_association/{llm_model}/{llm_model}_Unisex_gender_association_ALL.csv'
    df_personX = pd.read_csv(load_path_personX)
    df_unisex = pd.read_csv(load_path_unisex)
    plt.hist(df_personX["gender_association_score"], bins=30, alpha=0.5, label=f"{llm_model} (PersonX)")
    plt.hist(df_unisex["gender_association_score"], bins=30, alpha=0.5, label=f"{llm_model} (Unisex)")
    plt.xlabel("Gender Association Score")
    plt.ylabel("Word Count")
    plt.title("Distribution of Gender Association Scores")
    plt.legend()
    plt.show()
```

```python
plot_gender_associtation('bart')
```

```python
plot_gender_associtation('gpt2')
```
