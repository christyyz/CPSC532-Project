```python
import pandas as pd
import numpy as np

def gender_association_score(fileName, llm_model, neutral):
    # Load the lexical analysis CSV
    df = pd.read_csv(fileName)
    epsilon = 1e-6

    # Calculate Gender Association Score (GAS)
    df["gender_association_score"] = (
        (df["female_count"] - df["male_count"]) /
        (df["female_count"] + df["male_count"] + df["unisex_count"] + epsilon)
    )

    gas = df.reindex(df["gender_association_score"].abs().sort_values(ascending=False).index)

    # Save to CSV
    gas.to_csv(f'../results/{llm_model}_{neutral}_gender_association_score_ALL.csv', index=False)
```
